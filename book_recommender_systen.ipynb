# popularity based recommender system

```python
avg_rating=book_with_rating.groupby("Book-Title")["Book-Rating"].mean().reset_index()
```

```python
avg_rating=avg_rating.rename(columns={"Book-Rating":"avg_rating"})
```

```python
avg_rating
```

```python
num_rating=book_with_rating.groupby("Book-Title")["Book-Rating"].count().reset_index()
num_rating=num_rating.rename(columns={"Book-Rating":"num_rating"})
```

```python
num_rating
```

```python
popular_books=pd.merge(num_rating,avg_rating,on="Book-Title")
```

```python
popular_books=popular_books[popular_books["num_rating"]>200]
```

```python
popular_books
```

```python
popular_books=popular_books.sort_values(by="avg_rating",ascending=False).reset_index(drop=True)
```

```python
popular_books
```

```python
final_popular_df=pd.merge(popular_books,book_with_rating,on="Book-Title",how="left")
```

```python
final_popular_df
```

```python
final_popular_df["Book-Title"].duplicated().sum()
```

```python
final_popular_df.drop_duplicates("Book-Title",inplace=True)
```

# top 50 popular books

```python
top_30_popular_book=final_popular_df.head(30)
```

```python
x=book_with_rating.groupby("User-ID")["Book-Rating"].count() > 200
educated_users=x[x].index
```

```python
educated_users
```

```python
filtered_rating=book_with_rating[book_with_rating["User-ID"].isin(educated_users)]
```

```python
y=filtered_rating.groupby("Book-Title")["Book-Rating"].count()>50
famous_books=y[y].index
```

```python
final_df=filtered_rating[filtered_rating["Book-Title"].isin(famous_books)]
```

```python
pt=final_df.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating")
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarities=cosine_similarity(pt)
```

```python
similarities[0]
```

```python
def recommender(book_name):
    idx = np.where(pt.index == book_name)[0][0]
    similar_books = sorted(list(enumerate(similarities[idx])), key=lambda x: x[1], reverse=True)[1:6]
    
    for i in similar_books:
        idx=i[0]
        print(pt.index[idx])
```

```python
recommender("1984")
```

```python
import pickle
with open("similarities.pkl","wb") as f:
    pickle.dump(similarities,f)
```
